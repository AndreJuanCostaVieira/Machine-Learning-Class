## Criação de clusters "K Means"

Agora é hora de criar os rótulos de Cluster!

**Importe KMeans da SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

**Crie uma instância do modelo K Means com 2 clusters.**

```python
kmeans = KMeans(n_clusters=2)
```

```python
from sklearn.preprocessing import StandardScaler, normalize

scaler = StandardScaler() 
scaled_df = scaler.fit_transform(df.drop('Private',axis=1)) 
  
# Normalizando os dados
normalized_df = normalize(scaled_df) 
  
# Convertendo a matriz numpy em um DataFrame do pandas
normalized_df = pd.DataFrame(normalized_df) 
```

**Fite o modelo para todos os dados, exceto para o rótulo privado.**

```python
kmeans.fit(normalized_df)
```

**Quais são os vetores centrais do cluster?**

```python
kmeans.cluster_centers_
```

## Avaliação

Não há uma maneira perfeita de avaliar o agrupamento se você não tiver os rótulos, no entanto, como isso é apenas um exercício, temos os rótulos então aproveitamos isso para avaliar nossos clusters. Tenha em mente que não terá esse luxo no mundo real.

**Crie uma nova coluna para df chamado 'Cluster', que é 1 para escola particular e 0 para uma escola pública.**

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
    df['Cluster'] = df['Private'].apply(converter)
```

```python
normalized_df.head()
```

**Crie uma matriz de confusão e um relatório de classificação para ver o quão bem o clustering K Means funcionou sem ter nenhum rótulo.**

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```

```python
kmeans.predict
```

Não tão ruim, considerando que o algoritmo está usando apenas os recursos para agrupar as universidades em 2 grupos distintos. Espero que você possa começar a ver como K Means é útil para agrupar dados não rotulados!

```python
#vamos encontrar um k adequado
sse = {}
faixa = range(1,40)
for kclusters in faixa:
    kmeans1 = KMeans(n_clusters=kclusters)
    kmeans1.fit(df.drop('Private', axis=1))
    sse[kclusters] = kmeans1.inertia_ # Inertia: Soma das distâncias das amostras ao centro de aglomerado mais próximo

plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Número de cluster")
plt.ylabel("soma dos erros ao quadrado - SSE")
plt.title("Número de cluster por SSE - Algoritmo KMeans no SKLearn")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_scores = [] 

faixa = range(2,40)

for n_cluster in faixa:
    silhouette_scores.append( 
        silhouette_score(df.drop('Private', axis=1), KMeans(n_clusters = n_cluster).fit_predict(df.drop('Private', axis=1)))) 
    
# Plotando um gráfico de barras para comparar os resultados
k = list(faixa)
plt.bar(k, silhouette_scores) 
plt.xlabel('Número de clusters', fontsize = 10) 
plt.ylabel('Pontuação da silhueta', fontsize = 10) 
plt.show() 
```

```python
kmeans = KMeans(n_clusters=2, random_state=40)
kmeans.fit(df.drop('Private',axis=1))
print(classification_report(df['Cluster'],kmeans.labels_))
```
